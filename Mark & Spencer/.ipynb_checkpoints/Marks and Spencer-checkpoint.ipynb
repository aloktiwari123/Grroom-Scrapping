# Functions
## Function to get each product details

```python
def get_details(links_gender):
    try:
        main_link = links_gender[0]
        gender = links_gender[1]
        site_content = requests.get(main_link, headers=headers, cookies=cookies).content
        site_html    = bs(site_content, 'html.parser')
        if(site_html.find('div', {'class': 'productoutofstockmsg'})):
            print("Out of Stock")
            return None

        site_json = json.loads(site_html.findAll('script')[7].contents[0])
        card       = site_html.find("div", {"class": "product-detail-card"})
        amounts    = card.findAll("span", {"class": "value"})
        imgs       = eval(site_html.find("div",{'class':'image-handler-new-pdp new-pdp-carousal'})['data-product-images'])
        sizes      = card.find("select", {"class": "select-size"}).findAll("option")
        avail_size = ", ".join(i.text.strip() for i in sizes if i.text.strip().find("Stock")==-1 and i.text.strip().find("Select")==-1)

        product_details = {}
        product_details['link']          = main_link
        product_details['name']          = site_json['name']
        product_details['brand']         = bs(site_json['brand'], 'html.parser').text.strip()
        product_details['category']      = ", ".join(a.text.strip() for a in site_html.find('ol', {'class', 'breadcrumb'}).findAll("a"))
        product_details['avail_size']    = avail_size if len(avail_size)>0 else 'One Size'
        product_details['price']         = site_json['offers']['price']
        product_details['mrp']           = int(amounts[1]['content'].split(".")[0]) if 1<len(amounts) else product_details['price']
        product_details['gender']        = gender
        product_details['desc']          = bs(site_json['description'], 'html.parser').text.strip()
        product_details['primary_img']   = imgs[1]['url']+"?$PDP_IMAGEGRID_ZOOM_LG$"
        product_details['secondary_img'] = ", ".join(img['url']+"?$PDP_IMAGEGRID_ZOOM_LG$" for img in imgs[2:])
        return product_details

    except Exception as e:
        print(e, end="---")
        return None
```

# Create db

```python
conn = sqlite3.connect("marks_spencer.db")
cursor = conn.cursor()
try:
    cursor.execute("""CREATE TABLE product_details(
                    Website varchar(40) not null,
                    Product_Link text PRIMARY KEY,
                    Product_Name varchar(50) not null,
                    Product_Brand varchar(50) not null,
                    Product_Category varchar(50),
                    Size_Avail varchar(20) not null,
                    Price int not null,
                    MRP int not null,
                    Gender varchar(15) not null,
                    Description text not null,
                    Primary_Image_Links text not null,
                    Secondary_Image_Links text not null,
                    Affiliate_Link text
                )""")
    conn.commit()
except Exception as e:
    print(e)
    pass
```
